### Graphe 1 : Réaliser un graphique avec deux axes. L'axe des abscisses va de 0 à 50 et s'intitule "Axe des x". L'axe des ordonnées s'intitule "Axe des y".
##### Tracer la droite d'équation : x = 3y. Avec x allant de 0 à 50.
##### Le graphe possède le titre suivant "Tracer une droite" et la droite est de couleur rouge.

```python
# y = 3x est donc égal à 150 (je n'ai pas su faire autrement...)

# Je décide de l'axe des abcisses et des ordonnées du graphique.
x = [0.0,50.0]
y= [0.0,150.0]

#Je nomme les axes. 
plt.xlabel('Axe des x')
plt.ylabel('Axe des y')

# Je crée le graphique, avec une ligne rouge.
plt.plot(x, y, color="C3", linewidth =1,linestyle="-")

# Je l'affiche.
plt.show()
```

### Graphe 6
##### En prenant les 3 fichiers CSV étudiés précédemment, créer un graphique Pie Chart qui donne la répartition des produits par catégorie à afficher en pourcentage par produit.

```python
# Je lis les 3 documents CSV.
customers=pd.read_csv("/Users/claramoreschi/Desktop/TOUT DATA/27oct/customers.csv")
products=pd.read_csv("/Users/claramoreschi/Desktop/TOUT DATA/27oct/products.csv")
transactions=pd.read_csv("/Users/claramoreschi/Desktop/TOUT DATA/27oct/transactions.csv")

# Je fais un merge afin d'avoir un seul tableau, qui sera plus lisible.
data_final=pd.merge(transactions,products,how='left',on=["id_prod"]) 
data_final.shape
data_final=pd.merge(data_final,customers,how='left',on=["client_id"]) 

# Je n'affiche que les 5 premières lignes pour avoir de la visibilité.
data_final.head(5)
```

###### CLEAN

```python
# Dans la colonne date, je supprime les données qui ne me seront pas utiles.
new_date = data_final['date']
data_final['date'] = new_date.str[:19]

# Je crée deux nouvelles colonnes pour la date et l'heure.
data_final[['Date','Time']] = data_final["date"].str.split(' ',expand=True,)
data_final
```

```python
# Je supprime l'ancienne colonne date qui ne me servira plus.
del data_final['date']
```

```python
# J'affiche le tableau pour vérifier.
data_final
```
